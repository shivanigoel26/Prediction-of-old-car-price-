### Transforming data using function

#### New Price

```python
car_data.drop('New_Price',axis=1,inplace=True)
```

#### Engine

```python
# removing unit "CC" from Engine column to analyse data
def clean_CC(y):
    if isinstance(y, str):
        return(y.replace('CC', '').replace(',', ''))
    return(y)
# applying clean_CC() to values of engine column
car_data['Engine_CC'] = car_data['Engine_CC'].apply(clean_CC).astype('float')
```

#### Power

```python
# removing unit "bnp" from Power column to analyse data
def clean_bhp(z):
    if isinstance(z, str):
        return(z.replace('bhp', '').replace(',', ''))
    return(z)
# applying clean_bhp() to values of engine column
car_data['Power_bhp'] = car_data['Power_bhp'].apply(clean_bhp)
```

#### Mileage

```python
# Transforming data of mileage fuction
convert_dict1 = {'Mileage': str } 

car_data = car_data.astype(convert_dict1)

# function to chnage km/kg into kmpl
def kml(k):
    l = k.split(' ')
    if "km/kg" in k:
        return (str(float(l[0])) +" "+ "kmpl")
    if "kmpl" in k:
        return (str(float(l[0])) +" "+ "kmpl")
    else:
        return (str(float(l[0])))

# Applying Mileage column values in kml()
car_data['Mileage'] = car_data['Mileage'].apply(kml)

# removing units from  Mileage column to analyse data
def clean_unit(k1):
    if isinstance(k1, str):
        return(k1.replace('kmpl', '').replace(',', ''))
    return(k1)

# updating Mileage column values in data set
car_data['Mileage'] = car_data['Mileage'].apply(clean_unit)

#checking tha datatypes
car_data.dtypes
```

### Detecting outliers using IQR score

# Machine Learning Model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
```

```python
X=car_data[['Year','Engine_CC','Power_bhp','Kilometers_Driven','Fuel_Type','Transmission',
            'Owner_Type','Mileage','Seats','Manufacturer']]
y=car_data['Price_Lakh']
X_train, X_test, y_train, y_test = train_test_split( X,y,test_size = 0.3, random_state = 0)
print(len(X_train))
print(len(X_test))
```

```python
X_train = pd.get_dummies(X_train,
                         columns = ["Manufacturer", "Fuel_Type", "Transmission", "Owner_Type"],
                         drop_first = True)
```

```python
X_test = pd.get_dummies(X_test,
                         columns = ["Manufacturer", "Fuel_Type", "Transmission", "Owner_Type"],
                         drop_first = True)
```

```python
missing_cols = set(X_train.columns) - set(X_test.columns)
for col in missing_cols:
    X_test[col] = 0
X_test = X_test[X_train.columns]
```

```python
standardScaler = StandardScaler()
standardScaler.fit(X_train)
X_train = standardScaler.transform(X_train)
X_test = standardScaler.transform(X_test)
```

### Linear Regression

```python
linearRegression = LinearRegression()
linearRegression.fit(X_train, y_train)
y_pred = linearRegression.predict(X_test)
r2_score(y_test, y_pred)
```

### Random Forest Regression

```python
rf = RandomForestRegressor(n_estimators = 100)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
r2_score(y_test, y_pred)
```
