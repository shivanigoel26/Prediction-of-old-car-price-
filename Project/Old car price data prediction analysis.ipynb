# Transforming data usnig function

## New Price

```python
# converting data type of New_Price in str format 
convert_dict = {'New_Price_Lakh': str }
```

```python
car_data = car_data.astype(convert_dict)
```

```python
# function to chnage crore into lakh
def crore(x):
    l = x.split(' ')
    if "Cr" in x:
        return (str(float(l[0]) * 100) +" "+ "Lakh")
    if "Lakh" in x:
        return (str(float(l[0])) +" "+ "Lakh")
    else:
        return (str(float(l[0])))
```

```python
# Applying New_Price(in Lakh) column values in crore()
car_data['New_Price_Lakh'] = car_data['New_Price_Lakh'].apply(crore)
```

```python
# removing unit "Lakh" from column to analyse data
def clean_currency(x):
    if isinstance(x, str):
        return(x.replace('Lakh', '').replace(',', ''))
    return(x)
```

```python
# updating New_Price(in Lakh) column values in data set
car_data['New_Price_Lakh'] = car_data['New_Price_Lakh'].apply(clean_currency).astype('float')
```

## Engine

```python
# removing unit "CC" from Engine column to analyse data
def clean_CC(y):
    if isinstance(y, str):
        return(y.replace('CC', '').replace(',', ''))
    return(y)
```

```python
# applying clean_CC() to values of engine column
car_data['Engine_CC'] = car_data['Engine_CC'].apply(clean_CC).astype('float')
```

## Power

```python
# removing unit "CC" from Power column to analyse data
def clean_bhp(z):
    if isinstance(z, str):
        return(z.replace('bhp', '').replace(',', ''))
    return(z)
```

```python
# applying clean_bhp() to values of engine column
car_data['Power_bhp'] = car_data['Power_bhp'].apply(clean_bhp)
```

## Mileage

```python
# Transforming data of mileage fuction
convert_dict1 = {'Mileage': str } 
```

```python
car_data = car_data.astype(convert_dict1)
```

```python
# function to chnage km/kg into kmpl
def kml(k):
    l = k.split(' ')
    if "km/kg" in k:
        return (str(float(l[0])) +" "+ "kmpl")
    if "kmpl" in k:
        return (str(float(l[0])) +" "+ "kmpl")
    else:
        return (str(float(l[0])))
```

```python
# Applying Mileage column values in kml()
car_data['Mileage'] = car_data['Mileage'].apply(kml)
```

```python
# removing units from  Mileage column to analyse data
def clean_unit(k1):
    if isinstance(k1, str):
        return(k1.replace('kmpl', '').replace(',', ''))
    return(k1)
```

```python
# updating Mileage column values in data set
car_data['Mileage'] = car_data['Mileage'].apply(clean_unit)
```

# Identify and removing outliers

```python
#identify ouliers
for i in car_data.describe().columns:
    sns.boxplot(car_data[i])
    plt.show()
```

```python
#removing outliers
def remove_outliers(df):
    outliers={}
    for col in df.columns:
        if str(df[col].dtype)!='object':
            df=df[np.abs(df[col]-df[col].mean())<(3*df[col].std())]
            olrs=df[~(np.abs(df[col]-df[col].mean())<(3*df[col].std()))]
            outliers=pd.DataFrame(olrs)
    return df,outliers
```
